```python
%autoreload 2
from src import (
    eliminacion_gaussiana,
    descomposicion_LU,
    resolver_LU,
    matriz_aumentada,
    separar_m_aumentada,
)


# ####################################################################
def calc_determinante(A: list[list[float]]) -> float:
    """Funci√≥n que calcula el determinante usando el m√©todo...
    [Descomposici√≥n LU, eliminaci√≥n gaussiana, Gauss-Jordan, Gauss-Jacobi o Gauss-Seidel]

    ## Parameters
    ``A``: Matriz cuadrada de tama√±o n x n

    ## Return
    ``detA``: Determinante de la matriz A

    """
    # completar
    n = len(A)
    
    U = [row[:] for row in A]
    detA = 1
    for i in range(n):
        max_row = i
        for k in range(i + 1, n):
            if abs(U[k][i]) > abs(U[max_row][i]):
                max_row = k
        if i != max_row:
            U[i], U[max_row] = U[max_row], U[i]
            detA *= -1
        if U[i][i] == 0:
            return 0
        for k in range(i + 1, n):
            factor = U[k][i] / U[i][i]
            for j in range(i, n):
                U[k][j] -= factor * U[i][j]
    for i in range(n):
        detA *= U[i][i]
    return detA
# ####################################################################
    return
```
