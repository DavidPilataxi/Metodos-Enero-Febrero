2. Determine cuáles de las siguientes matrices son no singulares y calcule la inversa de esas matrices:  



```python
import numpy as np

# Definir matrices con nombres clave
matrices = {
    "A": np.array([[4, 2, 6], [3, 0, 7], [-2, -1, -3]]),
    "B": np.array([[1, 2, 0], [2, 1, -1], [3, 1, 1]]),
    "C": np.array([[1, 1, -1, 1], [1, 2, -4, -2], [2, 1, 1, 5], [-1, 0, -2, -4]]),
    "D": np.array([[4, 0, 0, 0], [6, 7, 0, 0], [9, 11, 1, 0], [5, 4, 1, 1]])
}

# Evaluar cada matriz
for key, mat in matrices.items():
    determinant = np.linalg.det(mat)

    if determinant:
        inverse_matrix = np.linalg.inv(mat)
        print(f"La matriz {key} es no singular (det ≠ 0).")
        print(f"Inversa de {key}:\n{inverse_matrix}\n")
    else:
        print(f"La matriz {key} es singular (det = 0), por lo que no tiene inversa.\n")
```

4. Encuentre los valores de A que hacen que la siguiente matriz sea singular.
  
  

```python
import sympy as sp

# Definir la variable simbólica
α = sp.Symbol('α')

# Definir la matriz A con α como parámetro
matrix_A = sp.Matrix([
    [1, -1, α],
    [2, 2, 1],
    [0, α, -3/2]
])

# Calcular el determinante de A
determinant_A = matrix_A.det()

# Encontrar los valores de α que hacen singular a la matriz (det = 0)
singular_values = sp.solve(determinant_A, α)

# Mostrar resultados
print(f"Determinante de A: {determinant_A}")
print(f"Valores de α que hacen que A sea singular: {singular_values}")
```

6. Factorice las siguientes matrices en la descomposición LU mediante el algoritmo de factorización *LU* con lii=1  para todas las i

  


```python
import sympy as sp

def lu_decomposition(mat, decimals=4):
    """Realiza la descomposición LU con l_ii = 1 para todas las i y trunca a 'decimals' decimales."""
    size = mat.shape[0]
    lower = sp.eye(size)
    upper = mat.copy()

    for row in range(size):
        for col in range(row + 1, size):
            multiplier = upper[col, row] / upper[row, row]
            lower[col, row] = multiplier
            upper[col, :] -= multiplier * upper[row, :]

    # Redondear matrices
    lower = lower.applyfunc(lambda x: round(x, decimals))
    upper = upper.applyfunc(lambda x: round(x, decimals))

    return lower, upper

# Definir las matrices
matrices = {
    "A": sp.Matrix([
        [2, -1, 1],
        [3, 3, 9],
        [3, 3, 5]
    ]),
    "B": sp.Matrix([
        [1.012, -2.132, 3.104],
        [-2.132, 4.096, -7.013],
        [3.104, -7.013, 0.014]
    ]),
    "C": sp.Matrix([
        [2, 0, 0, 0],
        [1, 1.5, 0, 0],
        [0, -3, 0.5, 0],
        [2, -2, 1, 1]
    ]),
    "D": sp.Matrix([
        [2.1756, 4.0231, -2.1732, 5.1967],
        [-4.0231, 6.0000, 0, 1.1973],
        [-1.0000, -5.2107, 1.1111, 0],
        [6.0235, 7.0000, 0, -4.1561]
    ])
}

# Aplicar factorización LU a cada matriz
for name, matrix in matrices.items():
    L, U = lu_decomposition(matrix)
    print(f"Factorización LU de la matriz {name}:")
    print(f"L:\n{L}\nU:\n{U}\n")
```

7. Modifique el algoritmo de eliminación gaussiana de tal forma que se pueda utilizar para resolver un sistema lineal usando la descomposición 
 y, a continuación, resuelva los siguientes sistemas lineales.  
 

```python
import numpy as np

def lu_decompose(matrix):
    """Descomposición LU con Lii = 1 para todas las i."""
    n = len(matrix)
    L = np.eye(n)  # Matriz L con diagonal de unos
    U = np.zeros((n, n))

    for i in range(n):
        for j in range(i, n):
            U[i, j] = matrix[i, j] - sum(L[i, k] * U[k, j] for k in range(i))
        
        for j in range(i + 1, n):
            L[j, i] = (matrix[j, i] - sum(L[j, k] * U[k, i] for k in range(i))) / U[i, i]

    return L, U

def forward_subst(L, b):
    """Sustitución progresiva para resolver Ly = b."""
    n = len(b)
    y = np.zeros(n)

    for i in range(n):
        y[i] = b[i] - sum(L[i, j] * y[j] for j in range(i))

    return y

def backward_subst(U, y):
    """Sustitución regresiva para resolver Ux = y."""
    n = len(y)
    x = np.zeros(n)

    for i in range(n - 1, -1, -1):
        x[i] = (y[i] - sum(U[i, j] * x[j] for j in range(i + 1, n))) / U[i, i]

    return x

def solve_system(A, b):
    """Resuelve el sistema Ax = b usando descomposición LU."""
    L, U = lu_decompose(A)
    y = forward_subst(L, b)
    x = backward_subst(U, y)
    return np.round(x, 4)  # Redondear la solución a 4 decimales

# Definir sistemas de ecuaciones
systems = {
    "a": {
        "A": np.array([
            [2, -1, 1],
            [3, 3, 9],
            [3, 3, 5]
        ]),
        "b": np.array([-1, 0, 4])
    },
    "b": {
        "A": np.array([
            [1.012, -2.132, 3.104],
            [-2.132, 4.096, -7.013],
            [3.104, -7.013, 0.014]
        ]),
        "b": np.array([1.984, -5.049, -3.895])
    },
    "c": {
        "A": np.array([
            [2, 0, 0, 0],
            [1, 1.5, 0, 0],
            [0, -3, 0.5, 0],
            [0, -2, 1, 1]
        ]),
        "b": np.array([3, 4.5, -6.6, 0.8])
    },
    "d": {
        "A": np.array([
            [2.1756, 4.0231, -2.1732, 5.1967],
            [-4.0231, 6.0000, 0, 1.1973],
            [-1.0000, -5.2107, 1.1111, 0],
            [6.0235, 7.0000, 0, -4.1561]
        ]),
        "b": np.array([17.102, -6.1593, 3.0004, 0])
    }
}

# Resolver cada sistema
for label, system in systems.items():
    A, b = system["A"], system["b"]
    solution = solve_system(A, b)
    print(f"Solución del sistema {label}: {solution}\n")
```
